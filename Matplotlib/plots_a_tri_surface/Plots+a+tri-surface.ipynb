```python
# Importar bibliotecas
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np

# Definir os parametros do grfico
n_angles = 36
n_radii = 8

# Uma matriz de radii
# No incluir o radius r=0, ele serve para eliminar pontos duplicados
radii = np.linspace(0.125, 1.0, n_radii)

# Uma matriz de angles
angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)

# Repetir todos os ngulos para cada radius
angles = np.repeat(angles[...,np.newaxis], n_radii, axis=1)

# Converte as coordenadas polares (radii, angles) para as coordenadas cartesianas (x, y)
# (0, 0). No h pontos duplicados no plano (x, y)
x = np.append(0, (radii*np.cos(angles)).flatten())
y = np.append(0, (radii*np.sin(angles)).flatten())

# Deefinir a superfcie
z = np.sin(-x*y)

# Utilizar a biblioteca figure para um output como imagem
fig = plt.figure()

#Definir que o grfico ser impresso em 3d
ax = fig.gca(projection='3d')
ax.plot_trisurf(x, y, z, cmap=cm.jet, linewidth=0.2)

#Imprimir os grficos
plt.show()
```
